```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
from wordcloud import WordCloud
```

```python
dataframe = pd.read_csv("../../Sample Datasets/tokenized_smsa_data.csv")
pd.set_option('display.max_colwidth', None)
dataframe.head(3)
```

```python
dataframe.info()
```

```python
df_sentiment = dataframe
```

```python
def get_sentiment(text):
    
    if isinstance (text, list):
        text = ' '.join(text)
        blob = TextBlob(text)
        sentiment = blob.sentiment.polarity
        
    if isinstance (text, str):
        blob = TextBlob(str(text))
        sentiment = blob.sentiment.polarity
        
        if sentiment >= 0.5 and sentiment <= 1:
            return "Positive"
        elif sentiment >= 0.05 and sentiment <= 0.49:
            return "Moderately Positive"
        elif sentiment >=  -1 and sentiment <= -0.5:
            return "Negative"
        elif sentiment >= -0.49 and sentiment <= -0.05:
            return "Moderately Negative"
        #elif sentiment >= -0.04 and sentiment <= 0.04:
            #return "Neutral"
        else: return "Neutral"
    else:
        return text
    
df_sentiment["sentiment"] =df_sentiment["Post"].apply(get_sentiment)
#df_sentiment.tail(3)
```

```python
sns.countplot(x = "sentiment", data = df_sentiment)
plt.savefig('blob.png')
#Neutral increases bcosof emojis' representation 
```

```python
def calc_polarity(text):
    blob = TextBlob(str(text))
    return blob.sentiment.polarity

df_sentiment["textblob_polarity"] = df_sentiment["Post"].apply(calc_polarity)
df_sentiment.head(3)
```

```python
def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color = 'white',
        max_words = 200,
        max_font_size = 40,
        scale = 3,
        random_state = 42
    ).generate(str(data))
    
    fig = plt.figure(1, figsize = (20, 20))
    plt.axis('off')
    if title:
        fig.suptitle(title, fontsize = 20)
        fig.subplots_adjust(top = 2.3)
    
    plt.imshow(wordcloud)
    plt.show()
    
show_wordcloud(df_sentiment["Post"])
```
